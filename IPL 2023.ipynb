```python
import pandas as pd
```

```python
df=pd.read_csv("ipl_2023_dataset.csv")
df
```

```python
df.head(244)
```

```python
index_name=df[df["2023 Squad"]=='Unsold'].index #returns index of Unsold values 
df=df.drop(index_name)
df
```

```python
# df['2022 Squad'].fillna("Did not play")
# df

# print(df['2022 Squad'][4])
df['2022 Squad'].fillna('Did not play',inplace=True)
# print(df['2022 Squad'][4])
df
```

```python
import matplotlib.pyplot as plt
bowler_2022=list(df[(df['Base Price']!="Retained") & (df['Type']=="BOWLER")]['2022 Squad'])
bowler_2023=list(df[(df['Base Price']!="Retained") & (df['Type']=="BOWLER")]['2023 Squad'])
# bowler_2022
# bowler_2023
player_name_2022=list(df[(df['Base Price']!="Retained") & (df['Type']=="BOWLER")]['Player Name'])
player_name_2023=list(df[(df['Base Price']!="Retained") & (df['Type']=="BOWLER")]['Player Name'])
# player_name_2022
plt.xticks(rotation=90)
plt.xlabel('Players')
plt.ylabel('Picked by team')
plt.title('Teams that picked non-retained bowlers-2022/2023')
plt.plot(player_name_2022,bowler_2022,label='2022')

plt.plot(player_name_2023,bowler_2023,label='2023')
plt.legend()

```

```python
bowler_price_2023=list(df[(df['Base Price']!="Retained") & (df['Type']=="BOWLER")]['Cost in Rs. (CR)'])
bowler_price_2023.sort(reverse=True)
print(bowler_price_2023)
plt.xticks(rotation=90)
plt.xlabel('Players')
plt.ylabel('Bought at cost(Rs)')
plt.title('Cost of non-retained players -2023')

plt.bar(player_name_2023,bowler_price_2023)
```
